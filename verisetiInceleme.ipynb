```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno

# Veri setini yükleyin
file_path = ("standardized_data.csv")
df = pd.read_csv(file_path)

# Veri setinin genel yapısına dair bilgi
print("Veri Seti İlk 5 Satır:")
print(df.head())

print("\nVeri Seti Bilgisi:")
print(df.info())

print("\nVeri Seti İstatistiksel Özeti:")
print(df.describe())

# Korelasyon Matrisi
correlation_matrix = df.corr()

# Korelasyon matrisini görselleştirme
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Korelasyon Matrisi')
plt.show()

# Verinin dağılımını görmek için histogram
plt.figure(figsize=(15, 12))
df.hist(figsize=(15, 12), bins=50)
plt.tight_layout()
plt.show()

# Hedef değişkenin (Class) dağılımı
sns.countplot(x='Class', data=df)
plt.title('Class Dağılımı (0: Normal, 1: Sahtekarlık)')
plt.show()

# Özellikler arasındaki ilişkiyi görmek için scatter plot (Örneğin: V13 ve V15)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='V13', y='V15', hue='Class', data=df, palette='coolwarm')
plt.title('V13 ve V15 Arasındaki İlişki (Class’a Göre)')
plt.show()

# Boxplot ile aykırı değerlerin analizi
plt.figure(figsize=(10, 6))
sns.boxplot(x='Class', y='V15', data=df)  # 'V15' özelliği örnek olarak seçilmiştir
plt.title('V15 Özelliği İçin Boxplot (Class’a Göre)')
plt.show()

# Eksik verileri görselleştirmek için missingno kullanarak eksik verileri gösterme
msno.matrix(df)
plt.show()
```
