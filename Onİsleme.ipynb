```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = ("augmented_fraud_data.csv")


df = pd.read_csv(file_path)


print(df.head())

print(df.info())

# Veri setinin temel istatistiksel zeti
print(df.describe())

# Korelasyon
correlation_matrix = df.corr()

# Korelasyon matrisini grselletirme
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Korelasyon eii
threshold = 0.7
to_drop = set()

# Korelasyon matrisinin sadece st geni
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            colname = correlation_matrix.columns[i]
            if colname != 'Class':  
                to_drop.add(colname)


df_cleaned = df.drop(columns=to_drop)

output_file_path = ("cleaned_fraud_data.csv")
df_cleaned.to_csv(output_file_path, index=False)

print("Korelasyonu yksek deikenler kaldrld ve yeni dosya kaydedildi: cleaned_fraud_data.csv")

# Veri setinin dalmn grmek iin histogram
df_cleaned.hist(figsize=(15, 12), bins=50)
plt.tight_layout()
plt.show()

# Hedef deikenin (Class) dalm
sns.countplot(x='Class', data=df_cleaned)
plt.title('Class Distribution (0: Not Fraud, 1: Fraud)')
plt.show()

# zellikler arasndaki ilikiyi grmek iin scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='V13', y='V15', hue='Class', data=df_cleaned, palette='coolwarm')
plt.title('Scatter Plot between V13 and V15 by Class')
plt.show()

# Boxplot ile aykr deerlerin analizi
plt.figure(figsize=(10, 6))
sns.boxplot(x='Class', y='V15', data=df_cleaned)  # 'V1' yerine baka bir zellik de seilebilir
plt.title('Boxplot of Feature V15 by Class')
plt.show()

# Eksik verileri grselletirmek iin missingno kullanabiliriz
import missingno as msno

# Eksik verileri grselletir
msno.matrix(df_cleaned)
plt.show()

# Veriyi normalize etmek iin Z-skoru kullanabiliriz
from sklearn.preprocessing import StandardScaler

# Z-skoru ile normalizasyon
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_cleaned.drop(columns=['Class']))

# Normalize edilmi veriyi grselletir
scaled_df = pd.DataFrame(scaled_features, columns=df_cleaned.columns[:-1])
scaled_df.hist(figsize=(15, 12), bins=50)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np

data = pd.read_csv("cleaned_fraud_data.csv")


data.drop(columns=['Time'], inplace=True)

data['log_amount'] = np.log1p(data['Amount'])  

data.drop(columns=['Amount'], inplace=True)

X = data.drop(columns=['Class']) 
y = data['Class'] 
scaler = MinMaxScaler()

X_scaled = scaler.fit_transform(X)

X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

X_scaled['Class'] = y.reset_index(drop=True)

print("Veri boyutu:", X_scaled.shape)
print(X_scaled.head())

X_scaled.to_csv("standardized_data.csv", index=False)
```
